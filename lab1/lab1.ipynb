```python
# ==========================================================
# STEP 1: Import all the required Python Libraries
# ==========================================================
import pandas as pd
import numpy as np


# ==========================================================
# STEP 2: Locate Open Source Dataset (URL)
# ==========================================================
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"


# ==========================================================
# STEP 3: Load Dataset into pandas DataFrame
# ==========================================================
# Dataset has no column names, so we provide them manually
columns = [
    "sepal_length",
    "sepal_width",
    "petal_length",
    "petal_width",
    "species"
]

df = pd.read_csv(url, header=None, names=columns)

print("First 5 Rows:")
print(df.head())


# ==========================================================
# STEP 4: Data Preprocessing
# ==========================================================

# Check dimensions
print("\nShape of dataset:")
print(df.shape)

# Check missing values
print("\nMissing values:")
print(df.isnull().sum())

# Statistical summary
print("\nStatistical Summary:")
print(df.describe())

# Column names
print("\nColumn names:")
print(df.columns)

# Data types
print("\nData types:")
print(df.dtypes)

# Variable types:
# Numeric -> sepal_length, sepal_width, petal_length, petal_width
# Categorical -> species


# ==========================================================
# STEP 5: Data Formatting and Normalization
# ==========================================================

# Convert species to category type
df['species'] = df['species'].astype('category')

print("\nAfter type conversion:")
print(df.dtypes)

# Normalize sepal_length using Min-Max scaling
df['sepal_length_norm'] = (
    df['sepal_length'] - df['sepal_length'].min()
) / (
    df['sepal_length'].max() - df['sepal_length'].min()
)

print("\nNormalized column:")
print(df[['sepal_length', 'sepal_length_norm']].head())


# ==========================================================
# STEP 6: Convert Categorical → Numerical using map()
# ==========================================================

df['species_numeric'] = df['species'].map({
    'Iris-setosa': 0,
    'Iris-versicolor': 1,
    'Iris-virginica': 2
})

print("\nCategorical to Numerical using map():")
print(df[['species', 'species_numeric']].head())


# ==========================================================
# FINAL OUTPUT
# ==========================================================
print("\nFinal Dataset:")
print(df.head())
```
