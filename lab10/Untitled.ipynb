```python
# ==========================================================
# STEP 1: Import Required Libraries
# ==========================================================
import pandas as pd
import matplotlib.pyplot as plt

# ==========================================================
# STEP 2: Load Dataset
# ==========================================================
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
df = pd.read_csv(url, header=None, names=columns)

# ==========================================================
# STEP 3: Display Dataset
# ==========================================================
print("First 5 Rows:")
print(df.head())

# ==========================================================
# STEP 4: Features and Their Types
# ==========================================================
print("\nFeature Types:")
print(df.dtypes)

# ==========================================================
# STEP 5: Histogram for Each Feature
# ==========================================================
df.hist(figsize=(10, 8))
plt.suptitle("Histogram of Iris Features")
plt.show()

# ==========================================================
# STEP 6: Boxplot for Each Feature
# ==========================================================
df.plot(kind='box', subplots=True, layout=(2,2), figsize=(10,8))
plt.suptitle("Boxplot of Iris Features")
plt.show()
```
