```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]

df = pd.read_csv(url, header=None, names=columns)

print(df.head())
```

```python
# Group by species and calculate statistics
grouped = df.groupby('species')['sepal_length']

stats = grouped.agg(['mean', 'median', 'min', 'max', 'std'])

print("Summary Statistics (sepal_length grouped by species):")
print(stats)

# Create list of values for each category
group_lists = grouped.apply(list)

print("\nList of values for each species:")
print(group_lists)
```

```python
# Separate datasets
setosa = df[df['species'] == 'Iris-setosa']
versicolor = df[df['species'] == 'Iris-versicolor']
virginica = df[df['species'] == 'Iris-virginica']

# Function to display stats
def display_stats(data, name):
    print(f"\nStatistics for {name}:")
    print("Mean:\n", data.mean(numeric_only=True))
    print("\nStandard Deviation:\n", data.std(numeric_only=True))
    print("\nPercentiles:\n", data.quantile([0.25, 0.5, 0.75], numeric_only=True))

# Call function
display_stats(setosa, "Iris-setosa")
display_stats(versicolor, "Iris-versicolor")
display_stats(virginica, "Iris-virginica")
```
